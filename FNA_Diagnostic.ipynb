```python
import pickle
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score

f=open('wdbc.pkl','rb')
data=pickle.load(f)
data.head()
Y = data['malignant']

X = data.drop(['malignant','id'],axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
data.tail()
```

```python
filter = (data["malignant"] == 0)

def printValue(df,col):
  print('mean,',df[col].mean())
  print('min,',df[col].min())
  print('max,',df[col].max())
print('Benign=>')
print('Area:')
printValue(data[filter],'area_0')
print('\nPerimeter_0:')
printValue(data[filter],'perimeter_0')
print('\nconcavity_0:')
printValue(data[filter],'concavity_0')

# data[filter]
```

```python
filter = (data["malignant"] == 1)

def printValue(df,col):
  print('mean,',df[col].mean())
  print('min,',df[col].min())
  print('max,',df[col].max())

print('Malignant=>')
print('Area:')
printValue(data[filter],'area_0')
print('\nPerimeter_0:')
printValue(data[filter],'perimeter_0')
print('\nconcavity_0:')
printValue(data[filter],'concavity_0')


# data[filter]
```

```python
from sklearn.ensemble import RandomForestClassifier


clf = RandomForestClassifier(n_estimators=200,n_jobs=-1)
clf = clf.fit(X_train, y_train)

# acc = accuracy_score(y_train, clf.predict(X_train))
# print('Training Accuracy: ', acc)

print('Test data report: ')
print(classification_report(y_test, clf.predict(X_test)))
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()

reg.fit(X_train, y_train)
print('Test data report: ')
print(classification_report(y_test, reg.predict(X_test)))
```
